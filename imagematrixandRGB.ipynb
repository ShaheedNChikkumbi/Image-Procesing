```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the image using OpenCV
input_image_path = r"C:\Users\Shaheed\Pictures\Camera Roll\WIN_20230228_13_36_08_Pro.jpg"  # Use a raw string
image = cv2.imread(input_image_path)
image_array = np.array(image)

print(image_array)

# Split the image into its RGB channels
blue_channel, green_channel, red_channel = cv2.split(image)

# Plot histograms for each channel
plt.figure(figsize=(15, 5))

# Blue channel histogram
plt.subplot(1, 3, 1)
plt.hist(blue_channel.ravel(), bins=256, color='blue', alpha=0.5)
plt.title('Blue Channel')
plt.xlim([0, 255])

# Green channel histogram
plt.subplot(1, 3, 2)
plt.hist(green_channel.ravel(), bins=256, color='green', alpha=0.5)
plt.title('Green Channel')
plt.xlim([0, 255])

# Red channel histogram
plt.subplot(1, 3, 3)
plt.hist(red_channel.ravel(), bins=256, color='red', alpha=0.5)
plt.title('Red Channel')
plt.xlim([0, 255])

plt.show()
```
